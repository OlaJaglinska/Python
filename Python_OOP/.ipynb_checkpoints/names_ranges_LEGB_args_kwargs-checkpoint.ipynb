- **Enclosing**

```python
def sample_function():
    
    var1 = 'Python'
    
    def inner_function():
        print(f'Variable var1: {var1}')
        print(f'Variable var2: {var2}')
        
    var2 = '3.8'
        
    inner_function()
```

```python
sample_function()
```

instruction 'nonlocal'

```python
def sample():
    
    num1 = 10
    
    def inner():
        nonlocal num1
        num1 += 20
        print(num1)
        
    print(num1)
    inner()
    print(num1)
    
sample()
```

**Exercises**

```python
counter = 1

def update_counter():
    global counter
    counter += 1
    print(counter)

    
update_counter()
```

```python
def display_info(number_of_updates=1):
    counter = 0
    dash_counter = ''
 
    def update_counter():
        nonlocal counter, dash_counter
        counter += 1
        dash_counter += '-'
        
    [update_counter() for _ in range(number_of_updates)]

    print(counter)
    print(dash_counter)
    
display_info(50)
```

```python
x = 10

def func1():
    globals()['x'] = 5
    
def func2():
    print(x)    

func1()
func2()
```

```python
x = 10

def func1():
    x = 5
    
    def inner_func():
        nonlocal x
        x = 7
    inner_func()
    print(x)

def func2():
    global x
    print(x)      
    
func1()
func2()
```

### * args **kwargs

```python
def sample(*args):
    print(len(args))
    print(args)
    
    
sample(4,5,6)
```

```python
def stick(*args):
    result = ''
    for arg in args:
        result += str(arg)
    return result

stick('Python', ' ', '3.8')
```

```python
def sample(**kwargs):
    print(len(kwargs))
    print(kwargs)
    
sample(var1='Python', var2=' 3.8')
```

```python
def func(**kwargs):
    for item in kwargs.items():
        print(item)
        
func(name='John', age='34')
```

```python
def hello(**kwargs):
    if 'name' in kwargs:
        print(f"Hello {kwargs['name']}!")
    else:
        print('Hello!')
        
hello()
```

```python
hello(name='Ann')
```

```python
def sum_int(**kwargs):
    if kwargs:
        result = 0
        for kwarg in kwargs.values():
            if isinstance(kwarg, int):
                result += kwarg
        return result
    return None
```

```python
sum_int(name='John', game1=34, game2=54, ratio=0.5)
```

```python
def sample(*args, **kwargs):
    print(args)
    print(kwargs)
    
sample(3, 5, 7, v1='py', var2='txt')
```

```python
stocks = {'appl':'Apple', 'tsla':'Tesla'}
sample(**stocks)
```

```python
numbers = (10, 20, 30)
sample(*numbers)
```

**Exercises**

```python
def stick(*args):
    args = [arg for arg in args if isinstance(arg,str)]
    result = ','.join(args)       
    return result
    
stick('sport', 'summer')
```

```python
stick(3, 5, 7)
```

```python
stick(False, 'time', True, 'workout', [], 'gym')
```

```python
def sum_numbers(*args):
    return sum(args)
    
sum_numbers(3,5)
```

```python
def info(main_tech, **techs):
    print(f"Main technology: {main_tech}")
    for k, v in techs.items():
        print(f'{k.upper()} -> {v}')
    
info('python', sql='junior')
```

```python
def concatenate_strings(delimiter = ',', **kwargs):
    values = [str(value) for value in kwargs.values()]
    return delimiter.join(values)
    
concatenate_strings(first_name='John', last_name='Doe', age=30)
```

```python
concatenate_strings()
```

```python
concatenate_strings('-', first_name='Bob', gender='Male', age=5)
```
