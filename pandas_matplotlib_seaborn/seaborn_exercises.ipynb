# Seaborn Exercises

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

**Recreate the plots below using the titanic dataframe.**

```python
sns.jointplot(x='fare', y='age', data=titanic, color='#2768FD', xlim=(-100,600))
```

```python
sns.distplot(titanic['fare'], color= 'red', kde=False, bins=30)
plt.xlim(0,600)
plt.ylim(0,500)
```

```python
sns.swarmplot(x='class', y='age', data=titanic, palette='pastel')
```

```python
sns.countplot(x='sex', data=titanic, palette='Set2')
```

```python
sns.heatmap(titanic.corr(), vmin=-1, vmax=1, cmap='coolwarm')
plt.title('titanic')
```

```python
g = sns.FacetGrid(data=titanic, col='sex')
g.map(plt.hist, 'age')
```
