```python
import pandas as pd
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3')
%matplotlib inline
```

**Recreate this scatter plot of b vs a.**

```python
df3.plot.scatter(x='a', y='b', s=60, c='#CD3720', figsize=(12,3), xlim=[-0.2, 1.2], ylim=[-0.2, 1.2])
```

**Create a histogram of the 'a' column.**

```python
df3['a'].plot.hist(color='blue', alpha=0.4)
```

**Use style sheets to set the style to 'ggplot' and redo the histogram from above. Also figure out how to add more bins to it.***

```python
plt.style.use('ggplot')
```

```python
df3['a'].plot.hist(color='red', alpha=0.5, bins=30)
```

**Create a boxplot comparing the a and b columns.**

```python
df3[['a','b']].plot.box()
```

**Create a kde plot of the 'd' column **

```python
df3['d'].plot.kde()
```

**Figure out how to increase the linewidth and make the linestyle dashed.**

```python
df3['d'].plot.kde(ls='dotted',lw=5)
```

** Create an area plot of all the columns for just the rows up to 30 and put legend outside.**

```python
df3[:31].plot.area(alpha=0.5)
plt.legend(loc='center',bbox_to_anchor=(1.1, 0.5))
```
