# Matplotlib Exercises 

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Exercise 1

* **Create a figure object called fig.**
* **Add an axis to the figure canvas at [0,0,1,1].**
* **Plot (x,y) on that axes and set the labels and titles.**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

## Exercise 2
* **Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**
* **Plot (x,y) on both axes.**

```python
fig2 = plt.figure()

ax1 = fig2.add_axes([0,0,1,1])
ax2 = fig2.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y, color='red')
ax2.plot(x,y, color='red')
```

## Exercise 3

* **Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x,z, color='green')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')


ax2.plot(x,y, color='darkblue')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim([20, 22])
ax2.set_ylim([30, 50])
```

## Exercise 4

* **Use plt.subplots(nrows=1, ncols=2) to create the plot.**

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
```

* **Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y, color='#75C111', ls='--', lw=3)
axes[1].plot(x,z, color='#FA8D11',lw = 4, alpha=0.7)

plt.tight_layout()
```

* **Try resize the plot.**

```python
fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(10,3))

axes[0].plot(x,y, color='#75C111', linestyle='--', linewidth=3)
axes[1].plot(x,z, color='#FA8D11',linewidth = 4, alpha=0.7)
```
